

### Insertion Traversal search and deletion

```python
class Node:
    
    # Function to initilize the node object
    def __init__(self,data):
        self.data = data # Assign data
        self.next = None # Initialize as null
        

# Linked List class contains a Node Object

class LinkedList:
    
    # Funtion to initialize head
    def __init__(self):
        self.head = None
        
    # Function to insert a new node at the beginning
    
    def push(self,new_data):
        
        # 1 & 2: Allocate the node & Put in the data
        new_node = Node(new_data)
        
        #3. Make next of new Node as head
        new_node.next = self.head
        
        #4. Move the head to point to new node
        self.head = new_node
        
    # FUnction to insert a node after the given Node
        
    def insertAfter(self,prev_node,new_data):
        # 1. Check if the given prev_node exists
        
        if prev_node is None:
            print("The given node must be in the linkedList")
            
            return
        
        # 2. Create new node and put in the data
        
        new_node = Node(new_data)
        
        # 3. Make the new node as next of prev Node
        new_node.next = prev_node.next
        
        # 4. Make next of prev_node as new_node
        prev_node.next = new_node
        
        
    ### Add new node in the end
    
    def append(self,new_data):
        
        # 1. Create new node
        # 2. Put in the data
        # 3. Set next as None
        
        new_node = Node(new_data)
        
        # 4 If the Liked List is empty, then make the new node as head
        
        if self.head is None:
            self.head = new_node
            return
        # 5. Else traverse till the last node
        last = self.head
        while (last.next):
            last = last.next
            
        # 6. Change the next of last node
        
        last.next = new_node
        
        
    ## Utility function to print the linked list
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data,end=" ")
            temp = temp.next
            
            
    ## Searching in linked List
    def search(self,value,key):
        
        # Base case
        if(not value):
            return False
        
        ## If key is present in current node, return true
        if (value.data == key):
            return True
        
        return self.search(value.next,key)
        
        
    def deleteNode(self,key):
        
        if self.head is None:
            return
        
        # Store head node
        temp = self.head 
        
        # If head node itself holds the key to be deleted
        if key is not None:
            if(temp.data == key):
                self.head = temp.next
                temp = None
                return
            
        # Search for the key to be deleted,keep track of the previous node
        # as we need to chagne 'prev.next'
        
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        

            
# Start with empty list

llist = LinkedList()

# Insert 5. So linked list become 6-> None
llist.append(6)

# Insert 7 at the beginning. So linked lisst becomes 7->6->None
llist.push(7)
       
# Insert 1 at thr beginning. So linked list becomes 1->7->6->None
llist.push(1)

# Insert 8 after 7. So linked list becomes 1->7->8->6-> None
llist.insertAfter(llist.head.next,8)

# Insert 10 after 8. So linked list becomes 1->7->8->6-> None
llist.insertAfter(llist.head.next.next,10)
```

```python
print('Created linked list is: ')
llist.printList()
```
